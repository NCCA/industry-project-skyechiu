# FaceLift Splat Results Viewer

Displays **input** / **multiview** / **output** for all splats in `data/splats/`.

```python
import os
from pathlib import Path
from PIL import Image
from IPython.display import display, HTML

SPLAT_DIR = Path(r'D:\zmm\facelift_pipeline\data\splats')

# Find all splat folders that have all 3 images
splat_folders = sorted([
    d for d in SPLAT_DIR.iterdir()
    if d.is_dir()
    and (d / 'input.png').exists()
    and (d / 'multiview.png').exists()
    and (d / 'output.png').exists()
])

print('Found ' + str(len(splat_folders)) + ' splat results')

import base64
from io import BytesIO

def img_to_base64(img_path, max_width=None):
    img = Image.open(img_path)
    if max_width and img.width > max_width:
        ratio = max_width / img.width
        img = img.resize((max_width, int(img.height * ratio)), Image.LANCZOS)
    buf = BytesIO()
    img.save(buf, format='PNG')
    return base64.b64encode(buf.getvalue()).decode()

for i, folder in enumerate(splat_folders):
    name = folder.name
    inp_b64 = img_to_base64(folder / 'input.png', max_width=256)
    mv_b64 = img_to_base64(folder / 'multiview.png', max_width=1200)
    out_b64 = img_to_base64(folder / 'output.png', max_width=1200)

    html = f'''
    <div style="border:1px solid #444; margin:10px 0; padding:10px; background:#1e1e1e;">
        <h3 style="color:white; margin:0 0 8px 0;">[{i+1}/{len(splat_folders)}] {name}</h3>
        <div style="display:flex; align-items:center; gap:10px;">
            <div style="text-align:center;">
                <div style="color:#aaa; font-size:12px;">Input</div>
                <img src="data:image/png;base64,{inp_b64}" style="height:200px;"/>
            </div>
            <div style="flex:1;">
                <div style="color:#aaa; font-size:12px;">Multiview</div>
                <img src="data:image/png;base64,{mv_b64}" style="width:100%;"/>
                <div style="color:#aaa; font-size:12px; margin-top:4px;">Output</div>
                <img src="data:image/png;base64,{out_b64}" style="width:100%;"/>
            </div>
        </div>
    </div>
    '''
    display(HTML(html))
```
